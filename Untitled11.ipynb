# PROFIT v/s DISCOUNT

```python
fig,ax=plt.subplots(figsize=(20,8))
ax.scatter(df['Sales'],df['Profit'])
ax.set_xlabel('Sales')
ax.set_ylabel('Profit')
plt.show()
```

```python
sns.lineplot(x='Discount', y='Profit', label='Profit', data=df)
plt.legend()
plt.show()
```

# PROFIT v/s QUANTITY

```python
sns.lineplot(x='Quantity', y='Profit', label='Profit', data=df)
plt.legend()
plt.show()
```

```python
df.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['green','blue'],figsize=(8,5))
plt.ylabel('Profit/Loss and Sales')
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.title('Segment based Sales for each Region')
sns.barplot(x='Region',y='Sales',data=df,hue='Segment',order=df['Region'].value_counts().index,palette='rocket')
plt.xlabel('Region',fontsize=14)
plt.ylabel('Sales',fontsize=14)
plt.show()
```

```python
df.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['orange','green'],figsize=(8,5))
plt.ylabel('Profit/Loss and Sales')
plt.show()
```

```python
df.groupby('State')[['Profit','Sales']].sum().sort_values(by='Sales',ascending=False).plot.bar(color=['red','blue'],figsize=(15,8))
plt.title('Profit/Loss and Sales across States')
plt.xlabel('State')
plt.ylabel('Profit/Loss and Sales')
plt.show()
```

```python
t_states=df['State'].value_counts().nlargest(10)
t_states
```

```python
df.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['darkblue','skyblue'],figsize=(8,5))
plt.ylabel('Profit/Loss and Sales')
plt.show()
```

```python
df.groupby('Sub-Category')[['Profit','Sales']].sum().sort_values(by='Sales',ascending=False).plot.bar(color=['orange','darkblue'],figsize=(15,8))
plt.title('Profit/Loss and Sales across States')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/Loss and Sales')
plt.show()
```
